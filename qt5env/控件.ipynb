```python
import sys

from PyQt5.QtWidgets import QWidget, QApplication, QLineEdit, QLabel
from PyQt5.QtCore import Qt


class Example(QWidget):

    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):
        # 常见标签、行编辑控件
        account = QLabel("1", self)
        account_line = QLineEdit(self)

        password = QLabel("2", self)
        password_line = QLineEdit(self)

        mail = QLabel("邮箱", self)
        mail_line = QLineEdit(self)

        # 设置位置，如果不设置就全部叠在一起了
        account.move(30, 15)
        password.move(30, 45)
        mail.move(30, 75)

        account_line.move(60, 10)
        password_line.move(60, 40)
        mail_line.move(60, 70)

        # 设置输出模式 密码输入模式
        # password_line.setEchoMode(QLineEdit.Password)
        # 设置掩码
        mail_line.setInputMask('xxxxxxxxxx@163.com; ')

        # 设置提示文本
        account_line.setPlaceholderText("请输入密码")
        password_line.setPlaceholderText("请输入用户名")

        self.setGeometry(300, 300, 400, 320)
        self.setWindowTitle('单行文本编辑器控件')
        self.show()


if __name__ == '__main__':
    app = QApplication(sys.argv)
    e = Example()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import (
    QApplication, QWidget, QTextEdit,
    QVBoxLayout, QPushButton
)

class Example(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):

        # 创建多行文本编辑器控件
        self.text = QTextEdit()
        # 创建两个按钮 控件
        self.show_text = QPushButton("显示文本")
        self.show_html = QPushButton("显示 HTML")
        self.show_image = QPushButton("显示图画")
        # 创建垂直布局管理器
        layout = QVBoxLayout()
        # 向垂直布局管理器中添加控件
        layout.addWidget(self.text)
        layout.addWidget(self.show_text)
        layout.addWidget(self.show_html)
        layout.addWidget(self.show_image)
        # 将布局管理器添加到 QWidget 的布局中
        self.setLayout(layout)

        # 给两个按钮空格键绑定点击槽函数（事件）
        self.show_text.clicked.connect(self.show_text_click)
        self.show_html.clicked.connect(self.show_html_click)
        self.show_image.clicked.connect(self.show_image_click)
        self.setGeometry(200, 200, 400, 320)
        self.setWindowTitle("多行文本编辑器控件")
        self.show()

    def show_text_click(self):
        """显示文本按钮控件槽函数"""

        # 设置文本
        self.text.setPlainText("欢迎来到蓝桥云课学习PyQt5!")

    def show_html_click(self):
        """显示 HTML 按钮控件槽函数"""

        # 设置 HTML
        self.text.setHtml("<p style='color:red;font-size:30px'>欢迎来到蓝桥云课学习PyQt5!</p>")
    def show_image_click(self):
        """显示 HTML 按钮控件槽函数"""

        # 设置 HTML
        self.text.setPlainText("<p style='color:red;font-size:30px'>欢迎来到蓝桥云课学习PyQt5!</p>")

if __name__ == "__main__":
    app = QApplication(sys.argv)
    e = Example()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QButtonGroup, QRadioButton

class Example(QWidget):
    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):
        # 创建按钮分组器控件
        group = QButtonGroup(self)

        # 创建单选按钮控件
        male = QRadioButton('男', self)

        # 设置“男”单选框被选中
        male.setChecked(True)
        male.move(150, 100)

        female = QRadioButton('女', self)
        female.move(200, 100)

        # 将两个按钮添加到按钮分组器中
        group.addButton(male)
        group.addButton(female)

        # 给单选按钮设置 ID
        group.setId(male, 1)
        group.setId(female, 2)

        # 打印所有按钮
        print(group.buttons())
        # 打印 ID=2 的按钮文本
        print(group.button(2).text())
        # 打印被按下的按钮文本
        print(group.checkedButton().text())

        self.setWindowTitle('按钮分组器控件')
        self.setGeometry(200, 200, 400, 320)
        self.show()
        print(group.checkedButton().text())
        self.show()
if __name__ == '__main__':
    app = QApplication(sys.argv)
    e = Example()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import (
    QApplication, QWidget, QTableWidget, QTableWidgetItem,
    QHeaderView
)
from PyQt5.QtCore import Qt


class Example(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):

        table = QTableWidget(4, 3, self)
        table.resize(420, 320)

        # 设置水平方向的表头标签，必须在表格控件初始化行列之后进行，否则，没有效果
        table.setHorizontalHeaderLabels(['姓名', '性别', '体重'])

        # 单元格水平方向自适应伸缩模式
        table.horizontalHeader().setSectionResizeMode(QHeaderView.Stretch)

        # 将单元格设置为禁止编辑模式
        table.setEditTriggers(QTableWidget.NoEditTriggers)

        # 设置表格整行选中
        table.setSelectionBehavior(QTableWidget.SelectRows)

        # 隐藏垂直表头
        table.verticalHeader().setVisible(False)

        # 添加数据
        newItem = QTableWidgetItem('张三')
        # 文本居中
        newItem.setTextAlignment(Qt.AlignCenter)
        table.setItem(0, 0, newItem)

        newItem = QTableWidgetItem('男')
        newItem.setTextAlignment(Qt.AlignCenter)
        table.setItem(0, 1, newItem)

        newItem = QTableWidgetItem('160')
        newItem.setTextAlignment(Qt.AlignCenter)
        table.setItem(0, 2, newItem)

        self.setWindowTitle("表格控件")
        self.setGeometry(200, 200, 420, 320)
        self.show()


if __name__ == '__main__':
    app = QApplication(sys.argv)
    e = Example()
    sys.exit(app.exec_())
```
